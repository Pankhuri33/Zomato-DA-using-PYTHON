```python
import pandas as pd

# Load dataset
zomato_data = pd.read_csv("/content/zomato.csv")
zomato_data.head()

```

```python
# Remove duplicates
zomato_data = zomato_data.drop_duplicates()

# Handle missing values
zomato_data = zomato_data.fillna(method='bfill')

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Example: Popular Cuisines
cuisines_count = zomato_data['cuisines'].value_counts()[:10]
sns.barplot(x=cuisines_count, y=cuisines_count.index)
plt.title("Top 10 Popular cuisines")
plt.show()
```

```python
top_locations = zomato_data['location'].value_counts().head(10)
sns.barplot(y=top_locations.index, x=top_locations)
plt.title("Top 10 Locations with Most Restaurants")
plt.xlabel("Number of Restaurants")
plt.ylabel("Locations")
plt.show()
```
